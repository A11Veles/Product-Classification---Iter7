### Combined dataset

```python
df = pd.read_csv(MWPD_FULL_DATASET_COMBINED)
df.head()
```

```python
df = df.applymap(
    lambda x: x.encode("ascii", "replace").decode("ascii") if isinstance(x, str) else x
)
```

```python
copy_to_sql(df, "mwpd_combined", "demo_user", if_exists="replace")
```

```python
copy_to_sql(df, "mwpd_combined_cleaned", "demo_user", if_exists="replace")
```

```python
cleaning_query = """
UPDATE demo_user.mwpd_combined_cleaned
SET Name = LOWER(
                  TRIM(
                    REGEXP_REPLACE(
                      REGEXP_REPLACE(
                        REGEXP_REPLACE(
                          REGEXP_REPLACE(Name, '[[:digit:]]+', ''), 
                          '[-_/\\|]', ' '
                        ),                              
                        '[[:punct:]]', ' '                              
                      ),
                      ' +', ' '   -- collapse multiple spaces into one
                    )
                  )
                ),SegmentTitle = LOWER(TRIM(
                          REGEXP_REPLACE(
                            REGEXP_REPLACE(
                              REGEXP_REPLACE(SegmentTitle, '[[:digit:]]+', ''), 
                              '[-_/\\|]', ' '),                              
                            '[[:punct:]]', ' '))),
                    FamilyTitle = LOWER(TRIM(
                          REGEXP_REPLACE(
                            REGEXP_REPLACE(
                              REGEXP_REPLACE(FamilyTitle, '[[:digit:]]+', ''), 
                              '[-_/\\|]', ' '),                              
                            '[[:punct:]]', ' '))),
                                ClassTitle = LOWER(TRIM(
                          REGEXP_REPLACE(
                            REGEXP_REPLACE(
                              REGEXP_REPLACE(ClassTitle, '[[:digit:]]+', ''), 
                              '[-_/\\|]', ' '),                              
                            '[[:punct:]]', ' '))),
                                BrickTitle = LOWER(TRIM(
                          REGEXP_REPLACE(
                            REGEXP_REPLACE(
                              REGEXP_REPLACE(BrickTitle, '[[:digit:]]+', ''), 
                              '[-_/\\|]', ' '),                              
                            '[[:punct:]]', ' '))),
                                Description = LOWER(TRIM(
                          REGEXP_REPLACE(
                            REGEXP_REPLACE(
                              REGEXP_REPLACE(Description, '[[:digit:]]+', ''), 
                              '[-_/\\|]', ' '),                              
                            '[[:punct:]]', ' ')));

"""
```

```python
td_db.execute_query(cleaning_query)
```

```python
tdf = td_db.execute_query("SELECT * FROM demo_user.mwpd_combined_cleaned;")
df = pd.DataFrame(tdf)
```
